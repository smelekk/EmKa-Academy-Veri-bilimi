```python
df = pd.read_csv('tips.csv')
```

```python
df.head()
```

```python
#Soru: ödenen ücreti 40dan büyük olan satırları getir 
df[df['total_bill'] > 40]
```

```python
#Soru: cinsiyeti erk2ek olan satırları
df[df['sex'] == 'Male']
```

```python
#Cinsiyetleri kadın ve bahsis miktarı 5ten fazla olansatırlar
df[(df['sex'] == 'Female') & (df['tip'] > 5)]    #pandasta and olmaz yerine& konur. ve parantezler olmak zorunda
```

```python
#Soru günü pazar olan ve ödenen miktar da 30dan fazla olsun
df[(df['day'] == 'Sun') & (df['total_bill'] > 30)]
```

```python
#Cumartesi veya pazar günü olan satırları getirin  | veya işareti
df[(df['day'] == 'Sat') | (df['day'] == 'Sun')]
```

**.isin metodu** 

```python
#cumartesi pazar pazartesi getirsin 
#tek tek yazmk yerine .isin ile yaparız -içinde mi???
df[df['day'].isin(['Sun','Sat','Thur'])] #tek df içinde yazınca True False gelir iki df ile yazdık   
```

**..dışında olanları nasıl isteriz?**  

```python
#sun dısındakiler
df[df['day'] != 'Sun']
```

```python
df[~(df['day'] == 'Sun')]  # ~ işareti de değil gibi kullanılıyor.
```
