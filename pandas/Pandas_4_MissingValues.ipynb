Missing valuelar farklı formatlarda olabilir.

#### Missing Valuelarla yapılabilecek 3 şey vardır.

```python
# 1-) Dokunmadan bırakabiliriz.
# 2-) Drop edebiliriz(satır veya sütun bazlı)
# 3-) Mantıklı bir şekilde doldurabiliriz.
```

```python
# 1. Yol: Yapılacak hiçbir şey yok. Bu şekilde işlemlerine devam et.
df
```

```python
# 2.Yol: 
df.dropna() #bir tane nan olsa bile drop eder
df.dropna(how='any') #ikisi aynı şey
```

```python
df.dropna(how= 'all') #heosi nan ise düşürür
```

```python
df.dropna(thresh= 1) #1tane bile nan olmayan veri varsa o satırı tut.
```

```python
df.dropna(thresh=4)
```

```python
df.dropna(thresh=5)
```

```python
df.dropna(axis=1) #sütun bazlı sorgulama yapar.tüm sütunlarda 1 tane nan olduğu için hepsini düşürdü
```

```python
df.dropna(axis=1, thresh=4) #thresh bir sütunun silinmemesi için geçerli min 4 verisi olmalı
```

```python
df
```

```python
# 3.Yol: Fill the data(Datayı doldur)
df.head()
```

```python

df.dropna(how='all', inplace=True) #hepsi nan olanı düşürdük kalanları dolduralım
```

```python
df.head()
```

```python
df['pre_movie_score'].fillna(9.5)
```

```python
df['pre_movie_score'].fillna(df['pre_movie_score'].mean()) #ortalama ile doldur
```
