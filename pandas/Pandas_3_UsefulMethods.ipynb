**.apply() Metodu**  
Bir series ya da dataframenin stunlarındaki değerlerin hepsine bir anda bir fonk uygulamaya yarar.  
Arka planında for döngüsü çalısır dolayısıyla pek hızlı değildir. Performans aşısından vektörize edilmiş işlemleri uygulamak daha faydalıdır.  
Özellikle vektörize edilmiş işlemler oldukça hızlı çalışır ve for döngüsü kullanımını minimize eder.  
Ancak diğer taraftan .apply() methodunu kullanmak oldukça okunaklıdır. Başlangıçta kolaydır ve her duruma uygulanabilir. Ancak vektörize işlemler bazı durumlarda yapılamayabilir.  
Satır satır dönüyor yani ilk satırı aldı inceledi 2.yi aldı inceledi 3. ..  


# AMAÇ: CC Number sütunundaki değerlerin son 4 rakamını almak ve CC Number sütununu bu şekilde güncellemek**
## 1.Yol

```python
# AMAÇ: CC Number sütunundaki değerlerin son 4 rakamını almak ve CC Number sütununu bu şekilde güncellemek 
# int bir şeyin son 4hanesini çekemeyiz o halde stinge döndürmeliyiz

def son4(x):
    return str(x)[-4:] #xi stringe döndür son 4 hanesini al

#artık fonk hazır bunu CC Number sütununa apply etmemiz gerekli
df['CC Number'] = df['CC Number'].apply(son4)
df.head() 
```

## 2.Yol

```python
df.head() # CC Number eski haline getirdik
```

```python
#fonksiyon oluşturmayacağız geçici fonk yapcaz
df['CC Number'] = df['CC Number'].apply(lambda x : str(x)[-4:]) #lambda kısa kullanım ve bellekte bosuna yer tutmaz 
df.head()
```

## 3.Yol(Vektörized)(En hızlısı bu yöntem)

```python
df.head()
```

```python
df['CC Number'].astype(str) #indexleme yapmak için string yaptık
#her satırın son 4ünü almak istiyoruz 
df['CC Number'].astype(str).str[-4:] # Vektörize edilmiş işlem str[] ile stringlerde işlem yapması için 
```

```python
df['CC Number'] = df['CC Number'].astype(str).str[-4:]
df.head()
```

#### Amaç: 10 dolardan az ödediyse $, 10-30 arası ile $$, 30dolardan fazla ödediyse $$$ işareti bulunan Fiyat Kategorisi adında yeni bir sütun oluşturmak

# 1.Yol

```python
def fiyat_kategori(x):
    if x < 10:
        return '$'
    elif 10<= x <30:
        return '$$'
    else:
        return '$$$'
```

```python
df['Fiyat Kategorisi'] = df['total_bill'].apply(fiyat_kategori)
df.head()
```

# 2.Yol

```python
df.head()
```

```python
df['Fiyat Kategorisi'] = df['total_bill'].apply(lambda x: '$' if x < 10 else('$$' if 10 <= x < 30 else '$$$'))
df.head()
```

# 3.Yol Vektorized(En hızlısı)  
tek tek satır dönerek incelemk yerine bunu kulanırız daha hızlıdır

```python
df.head()
```

```python
df.drop('Fiyat Kategorisi', axis = 1, inplace=True)
```

```python
df['total_bill']
```

```python
def fiyat_kategori_vectorized(bill):   #bütün sütun olarak alır total_billi
    categories = np.full_like(bill, '$', dtype='U3')             #full_like şunun gibi doldur demek 
    
    categories[(bill >= 10) & (bill < 30)] = '$$'
    
    categories[bill >= 30] = '$$$'
    
    return categories
```

```python
df.head()
```

```python
fiyat_kategori_vectorized(df['total_bill'].values)
```

```python
#bunu yeni bir sütun olarak dosyaya ekleyelim 
df['Fiyat Kategori'] = fiyat_kategori_vectorized(df['total_bill'].values)
df.head()
```

```python
df.sample(10)
```
