```python
import numpy as np
import pandas as pd
```

### Korelasyon nedir?  
-1.0 ile 1.0 arası bir sayıdır. Değerlerin birbiriyle olan ilişkisine bakıyor.  
+1.0 iki değişkenin aynı yönde hareket ettiğini gösterir yani bir iartarken diğeri de artar  
-1.0 ise zıt yönde hareket ettğiğini gösterir yani biri artarken diğeri azalır 

```python
df = pd.read_csv('mpg.csv')
```

```python
# 1) df'in içindeki sütunların birbirleriyle olan korelasyonlarına(ilişkilerine) bakın.
korelasyon = df.corr(numeric_only = True)
korelasyon
```

```python
# 2) mpg sütunu ile diğer sütunların arasındaki korelasyonları özel olarak inceleyin.
# (Bunu göstermedik ama df.corr() çağırdıktan sonra aynı normal df'lerde olduğu
# gibi istediğiniz sütunu çağırabiliyorsunuz.)
korelasyon['mpg']
```

```python
# 3) cylinders sütununda hangi unique değerler olduğuna liste halinde bakın.
df['cylinders'].unique()
```

```python
# 4) cylinders sütunundaki unique değerlerin her birinden kaçar adet olduğuna bakın.
df['cylinders'].value_counts()
```

```python
df.head(3)
```

```python
# 5) Arabanın modeli 75-80 yılları arasındaki arabaların hepsini getirin.(75 ve 80 dahil)
df[df['model_year'].between(75,80)]
```

```python
# 6) horsepower'ı 155-180 arasındaki arabaların hepsini getirin.(155 dahil, 180 dahil değil)
# (Bu sorguyu yapabilmeniz için horsepower sütununu object durumdan float duruma getirmeniz gerekiyor.)
# '?' karakterini NaN (Not a Number) olarak değiştirme
df['horsepower'] = df['horsepower'].replace('?', pd.NA)
df['horsepower'] = df['horsepower'].dropna().astype(int)
arabalar = df['horsepower'].between(155,180,inclusive='left')
df[arabalar]
```

```python
# 7) Elinizdeki df'i horsepower'a göre büyükten küçüğe sıralayın ardından 
# df'in indexlerini sıfırlayın.
sıralı = df.sort_values(by='horsepower', ascending=False).reset_index() #ascending=Fasle azalan sırada 
sıralı
```

```python
# 8) Elinizdeki df'den en küçük 5 acceralation'a sahip olan arabaları getirin  
# --- CIA_Country_Facts.csv dosyasını okuyun ve yeniden df değişkenine atayın.
# (Artık eski df ile işimiz kalmadığı için aynı ismi verebiliriz.)
df.sort_values(by='acceleration').head(5)
```

```python
df = pd.read_csv('CIA_Country_Facts.csv')
df.head(3) 
```

```python
# 9) Elimizdeki CIA'in ülkeler hakkında topladığı basit açık kaynak verileri df'inde
len(df)
df.shape[0]
```

```python
# 10) Elimizdeki sütunlardan hangilerinde kaçar tane nan veri olduğunu bulun.
df.isna().sum()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
# 11) Elimizde tamamen nan değerlerden oluşan bir satır varsa o satırı düşürün.
# (Tamamen nan değerlerden oluşan bir satır olup olmadığını, herhangi bir satırın 
# düşüp düşmediğini kontrol edin)
len(df.dropna(how='all')) 
#o zaman tamamı nan değerde noluşan satır yok. Hiçbir satır düşürülmemiş
```

```python
# 12) Her Bölgedeki(Region) toplam kişi sayısına bakın.
df['Region'].value_counts()
```

```python
#veya bu şekilde de çözülebilir ama üstteki daha hızlı 
df.groupby('Region').size()  #groupby ile gruplara ayırdık boyutuna baktık 
```

```python
df.head()
```

```python
# 13) Climate değerleri nan olanlar, tüm df'in ortalama climate değeriyle doldurun.
# (Gerçek bir analiz yapıyor olsaydık bu çok mantıklı bir hareket olmazdı, belki her 
# bulunduğu bölgenin ortalama climate değeriyle doldurmak daha mantıklı bir hareket olurdu. Fakat şimdilik bu kadar kasmaya gerek yok. 
# nan değerleri tüm climate ortalamalarıyla doldursanız yeterli.)
df['Climate'].fillna(df['Climate'].mean())
df.sample(3)
```

```python
df.head(10)
```

```python
# 14) Bölgeler ve Ülkelere göre gruplama yapın ardından Doğum miktarlarına bakın.
istenen = df.groupby(by=['Region', 'Country'])['Birthrate'].mean()
istenen
```

```python
# 15) Veriyi istediğiniz şekilde inceleyerek bu zamana kadar öğrendiğiniz bilgilerle 
# anlamlı 5 adet yargı cümlesi çıkartın. ( En çok insan Asya kıtasında yaşıyor,
# En çok insan Afrinın sahralarında ölüyor, Ortalama en yüksek telefon sahibi Kuzey Amerikada, 
# En çok okuma yazma oranına sahip ülkeler Baltık ülkeleri vb.)
istenen.index.get_level_values('Region').value_counts()
```

```python
#diğer bölgelere göre en çok tarıma dayalı bölgeleri bul 
yüzde = df.groupby(by='Region')['Agriculture'].mean()*100 #her bölgedeki % tarım payı 
yüzde
```

```python
#yüzde 10 üstündekielr tarıma dayalı olsun onları bulalım 
elverisli = yüzde[yüzde>= 10]
elverisli
```

```python
#sektörlerde Ekonomik bağımlılık hangi ülkeler içn daha önemli 
# Agriculture GSYİH tarım, Industry GSYİH sanayi, Service GSYİH  hizmet(Sayısal)
ülke_satırı_tarımda = df[df['Agriculture'] == df['Agriculture'].max()]
ülke_satırı_tarımda
print(f"tarım payının en yükske olduğu ülke: {ülke_satırı_tarımda['Country'].iloc[0]} %{(ülke_satırı_tarımda['Agriculture'].iloc[0]*100):.2f}")
```

```python
ülke_satırı_sanayide = df[df['Industry'] == df['Industry'].max()]

print(f"sanayi payının en yükske olduğu ülke: {ülke_satırı_sanayide['Country'].iloc[0]} %{(ülke_satırı_sanayide['Industry'].iloc[0]*100):.2f}")
```

```python
ülke_satırı_hizmet = df[df['Service'] == df['Service'].max()]

print(f"hizmet payının en yükske olduğu ülke: {ülke_satırı_hizmet['Country'].iloc[0]} %{(ülke_satırı_hizmet['Service'].iloc[0]*100):.2f}")
```
