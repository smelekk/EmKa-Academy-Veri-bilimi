```python
import numpy as np
import pandas as pd
```

```python
df_tips = pd.read_csv("tips.csv")
df_tips.head()
```

```python
# 1) df_tips içindeki Payment ID column'unun son 3 rakamını çekin. Bu işlemi 3 farklı yolla da yapmaya çalışın.
# 1.yol fonksiyon+apply()
def son3(x):
    return str(x)[-3:]
df_tips['Payment ID'].apply(son3)
```

```python
# 2.yol lambda+apply()
df_tips['Payment ID'].apply(lambda x: str(x)[-3:])
```

```python
# 3.yol vektörized hızlı olan
df_tips.info()  #object diyor 
df_tips['Payment ID'].astype(str).str[-3:]
```

```python
df_tips.head(3)
```

```python
# 2) df_tips içindeki Payer Name sütununda isim ve soy isim bilgileri bulunuyor.
# O sütundan sadece isim bilgileri çekin ve 'JustName' adında yeni bir column oluşturun.
# (Bu biraz zorlayıcı olabilir, Önce kendiniz yapmaya çalışın fakat yapamazsanız ChatGPT'ye 
# soruyu düzgünce sorarak cevabını kod olarak alabilirsiniz, ardından döndürülen kodu anlamaya çalışmanız yeterlidir.)
df_tips['just_name'] = df_tips['Payer Name'].str.split(' ').str[0]
df_tips
 
```

```python
# 3) 2.Soruda yeni oluşturduğunuz 'JustName' sütunundaki verilerin her 
# birinden kaç tane olduğuna uygun methodu kullanarak bakın.
df_tips['just_name'].value_counts()
```

```python
# 4) 3.Soruda gördüğünüz gibi JustName sütununda 'Jason' isminden 9 tane, 
# 'Michael' isminden 7 tane , 'James' isminden 6 tane bulunuyor.
# Peki bunların hepsi gerçekten farklı kişiler mi ? Hemen 'Payer Name' sütunundaki
# verilerin adetlerini de saydırtarak bir kıyaslama yapın. 'JustName' sütununu kullanmanın
#  faydasız ve yanıltıcı olabileceğini görün ardından oluşturduğumuz 'JustName' sütununu kalıcı olarak drop edin. 
#  (Oluşturmanız için bir pratikti bu sütun sadece :D)
df_tips['Payer Name'].value_counts()
```

```python
df_tips.drop('just_name', axis=1,inplace=True)
```

```python
df_tips.head()
```

```python
# 5) Erkeklerin ve kadınların davranış patternlerini ayrı ayrı inceleyip yorumda
# bulunabilmek için 'df_male' ve 'df_female' adında 2 dataframe oluşturun.
# Birinde sadece 157 erkek verisi, diğerinde sadece 87 kadın verisi bulunsun.
df_male = df_tips[df_tips['sex'] == 'Male']
df_male = df_male.sample(157)
df_female = df_tips[df_tips['sex'] == 'Female']
df_female = df_female.sample(87)
df_male.reset_index()
df_female.reset_index()
```

```python
# 6) Ardından bu 'df_male' ile 'df_female' 'in ödedikleri bahşişler toplamını bulup karşılaştırın.
male_toplam = df_male['tip'].sum()
femal_toplam = df_female['tip'].sum()

print(df_male['tip'].sum())
print(df_female['tip'].sum())

#yüzde erkekler kadınlardan kaç fazla ödemiş 
yüzde = ((male_toplam-femal_toplam)/femal_toplam)*100
yüzde
```

```python
# 7) 6.Soruda erkeklerin ve kadınların ödedikleri bahşişler toplamına bakıldığında Erkeklerin kadınlara kıyasla
# toplamda daha çok bahşiş ödediğini görüyoruz. Fakat biz biliyoruz ki bizim verimizde 157 Erkek, 87 kadın 
# verisi vardı. Dolayısıyla Erkek sayısı daha fazla olduğu için toplam ödeyecekleri bahşiş miktarlarında da 
# daha fazla çıkması oldukça normal bir durum. Bu yüzden daha mantıklı bir çıkarım için bu sefer 
# df_male ile df_female'in bahşiş ORTALAMALARIN'a bakıp kıyaslayın, aralarında anlamlı ve ciddi bir fark var mı ?
male_ort = df_male['tip'].mean()
femal_ort = df_female['tip'].mean()

print(df_male['tip'].mean())
print(df_female['tip'].mean())
#yüzdelik olarak erkekler kadınlara göre kaç fazla?
yüzde= ((male_ort - femal_ort)/femal_ort)*100
yüzde
```

```python
# 8) Oluşturmuş olduğunuz df_male ile çalışarak elimizdeki verideki erkeklerin
# yüzde kaçının sigara içtiğini bulun. Aynı şekilde elimizdeki kadınların yüzde
# kaçının sigara içtiğini de bulun. Aralarında anlamlı bir fark mı yorumlayın.

df_male_sigara_sayi = len(df_male[df_male['smoker'] == 'Yes'])
df_female_sigara_sayi = len(df_female[df_female['smoker'] == 'Yes'])
male_yüzde = (df_male_sigara_sayi/ (len(df_male)))*100
female_yüzde = (df_female_sigara_sayi/ (len(df_female)))*100
print(f"kadın yüzdesi: {female_yüzde}")
print(f"erkek yüzde: {male_yüzde}")
```

```python
kadin_sigara_orani = df_female['smoker'].value_counts(normalize=True)*100
erkek_sigara_orani = df_male['smoker'].value_counts(normalize=True)*100
print(kadin_sigara_orani)
erkek_sigara_orani
```

```python
# 9) Erkeklerin zaman olarak dinnerda mı yoksa lunch'da mı daha fazla yemek 
# yediğini yüzdesel olarak bulun. Aynı işlemi kadınlar için de yapın.
# Yüzdesel olarak erkeklerin ve kadınların yemek yeme vakitleri arasında anlamlı bir fark var mı yok mu yorumlayın.
erkek_yemek = df_male['time'].value_counts(normalize = True) *100
kadın_yemek = df_female['time'].value_counts(normalize = True) *100
print(f"erkek yüzdeseli: {erkek_yemek}")
print(f"kadın yüzdeseli: {kadın_yemek}")
```

```python
# 10) CC Number sütununu kalıcı olarak drop edin. Ardından df'in içindeki
# sütunların birbirleriyle olan ilişkilerini inceleyin ve çıkan sonuçlar hakkında yorumlarınızı belirtin.
df_tips.drop('CC Number', axis =1, inplace =True)
```
