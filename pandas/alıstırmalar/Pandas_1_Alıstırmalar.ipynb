```python
import numpy as np
import pandas as pd
```

```python
# 1) İndexlerinde "Breaking Bad" , "Chernobyl" , "Leyla ile Mecnun" , "X-Men"  , "Captain America"  , "Disaster Movie" olan 
#    Datasında 9.4 , 9.3 , 8.8 , 8.4 , 3.8 , 2.1 olan bir seri oluşturun.(2 Farklı yolla da oluşturun) 
# (Datayı ve İndexleri liste olarak göndererek ve doğrudan dictionary göndererek.)
index = [ "Breaking Bad" , "Chernobyl" , "Leyla ile Mecnun" , "X-Men"  , "Captain America"  , "Disaster Movie"]
data = [9.4 , 9.3 , 8.8 , 8.4 , 3.8 , 2.1]
pd.Series(data = data, index = index) 
```

```python
'''
2) donem1_mat adında bir seri oluşturun. Bu serinin içinde "Ali" ,"Deniz" ,"Aslı" ,"Burak" ,"Kerem" ,"Yasemin" ,"Derya" 
index değerleri olsun,85, 80 , 94 , 74 , 91 , 96 , 70 data değerleri olsun.
Bu şekilde donem1 adında öğrencilerin 1.dönem matematik nor ortalamalarının
tutulduğu bir seri oluşturun.donem2_mat adında başka bir seri oluşturun.
Okula yeni bir öğrenci gelmiş olsun ve indexlerde bu sefer "Ali" ,"Deniz" ,"Aslı" ,"Burak" ,"Kerem" ,"Yasemin" ,"Derya" ,"Mert"
indexleri olsun. 80, 92 , 85 , 87, 90 , 92, 100, 83 ise dataları olsun. 
Bu iki seriyi birleştirerek bütün öğrencilerin yıl sonu matematik ORTALAMASINI hesaplayın.
Not : Eğer bir öğrencinin sadece ikinci dönem notu bulunuyorsa, birinci dönem hesaba eklenmeden
ikinci dönem notu yıl sonu notu olarak hesaplansın.
'''
isim = [ "Ali" ,"Deniz" ,"Aslı" ,"Burak" ,"Kerem" ,"Yasemin" ,"Derya" ]
notlar = [85, 80 , 94 , 74 , 91 , 96 , 70]
donem1_mat = pd.Series(notlar, isim)
isim_2 = ["Ali" ,"Deniz" ,"Aslı" ,"Burak" ,"Kerem" ,"Yasemin" ,"Derya" ,"Mert"]
notlar_2 = [80, 92 , 85 , 87, 90 , 92, 100, 83]
donem2_mat = pd.Series(notlar_2, isim_2)
df = pd.DataFrame({'donem 1': donem1_mat, 'donem 2': donem2_mat})
ort = df.mean(axis =1)
print(ort)
df
```

```python
# 3) İndexlerinde "Osman", "Ozlem", "İnci", "Koray", "Melek" olan, 
# columnlarında "Sınıf" , "Cinsiyet" , "Favori Ders" isimlerinde 3 column bulunan 
# bir DF oluşturacaksınız. 
# Osman'ın sınıfı 8 , cinsiyeti "E" , Favori Dersi "Fen"
# Ozlem'in sınıfı 11 , cinsiyeti "K" , Favori Dersi "Mat"
# İnci'nin sınıfı 5 , cinsiyeti "K" , Favori Dersi "Sosyal"
# Koray'ın sınıfı 9 , cinsiyeti "E" , Favori Dersi "Mat"
# Meleğin sınıfı 7 , cinsiyeti "K" , Favori Dersi "Ing" 
# Olacak şekilde bir df oluşturun.
veri = {
    "Sınıf": [8, 11, 5, 9, 7],
    "Cinsiyet": ["E", "K", "K", "E", "K"],
    "Favori Ders": ["Fen", "Mat", "Sosyal", "Mat", "İng"]
}

indexler = ["Osman", "Özlem", "İnci", "Koray", "Melek"]
df = pd.DataFrame(data =veri, index = indexler)
df
```

```python
# 6) df_adv içinden sadece "TV" sütununu seri olarak çekin.
df[['TV']]
```

```python
# 7) df_adv içinden "TV" ve "sales" sütununu çekin ve df_salesTV adında bir değişkene atayın.
df_salesTV = df[['TV', 'sales']]
df_salesTV
```

```python
# 8) df_adv içindeki "TV" , "radio" , "newspaper" sütunlarındaki değerleri toplayarak
#  "total_adv" adında yeni bir sütun oluşturun.
sütunlar = ['TV', 'radio', 'newspaper']
df[total_adv] = df[sütunlar].sum(axis=1)
total_adv
```

```python
# 9) df_adv içindeki "sales" sütununun drop edilmiş halini ekrana yazdırın.
#  (Kalıcı hale getirmeyin. Sadece ekrana bastırın)
df_adv.drop('sales',axis = 1)
```

```python
# 10) df_adv içindeki son 50 index'i(satırı) çekin ve ekrana yazdırın.
df_adv.tail(50)
```

```python
# 11) df_adv içinde satışları 25'den fazla olanları getirin.
df_adv[df_adv['sales'] > 25]
```

```python
# 12) df_adv içinde "TV" reklamlarına 250'den fazla VE "newspaper" reklamlarına 60'dan fazla harcama yapanları getirin.
df_adv[ (df_adv['TV'] >250) & (df_adv['newspaper'] > 60)]
```

```python
# 13) df_adv içinde "newspaper" reklam harcamaları 70'den fazla VEYA
#  "radio" reklam harcamaları 50'den fazla olanları getirin.
df_adv[ (df_adv['radio'] >50) | (df_adv['newspaper'] > 70)]
```

```python
# 14) df_adv içinde "sales" ları 10 ile 15(10 ve 15 dahil) arasında olanları getirin
df_adv[(10 < df_adv['sales'])& (df_adv['sales'] <= 15)]
```

```python
# 15) df_adv içinde "sales" ları tam olarak 18.3 veya 14.0 veya 
# 25.5 veya 21.7 veya 13.6 veya 6.9 olanları getirin.

df_adv[df_adv['sales'].isin([25.5, 21.7, 13.6, 6.9])]
```

```python
# 16) GitHubda bulunan "tips.csv" restoran datasını "df_tips" adıyla okuyun.
#  Ardından ilk 5 satırına bakın.
df_tips = pd.read_csv("tips.csv")
df_tips.head()
```

```python
# 17) df_tips'in "CC Number" ve "Payment ID" column'larını kalıcı olarak düşürün.
df_tips.drop(columns=["CC Number", "Payment ID"], inplace = True)
```

```python
# 18) df_tips'in "Payer Name" column'unu kalıcı olarak index yapın ve artık columndan düşürün.
df_tips.set_index('Payer Name', inplace=True)
```

```python
# 19) df_tips içindeki kayıtlardan kişi sayısı 4 veya 4'den fazla olan siparişlerin 
# kaç tane olduğunu bulun.(Sorguyu yazıp istediğimiz satırları getirdikten sonra 
# basitçe len() fonksiyonunu kullanarak sayılarını bulabilirsiniz. Veya daha güzeli 
# .shape[0] yaparak elde ettiğiniz sorguda kaç satır olduğunun sayısına ulaşabilirsiniz.)
sayi = df_tips[df_tips['size'] >= 4]
print(len(sayi))
sayi.shape[0]
```

```python
# 20) df_tips içindeki Kişi sayısı 2 ya da 3 olan, sigara içen, erkek olan
#  ve ödediği ücret 20 dolardan az olan kişi sayısını bulalım.
kosullu_kisiler = df_tips[(df_tips['size'].between(2,3)) & (df_tips['smoker'] == 'Yes') & (df_tips['sex']=='Female') & (df_tips['total_bill'] < 20)]
len(kosullu_kisiler)
```
